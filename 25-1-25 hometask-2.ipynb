```python
import pandas as pd
df=pd.read_csv("C:\\Users\\vani1\\Desktop\\r\\GlobalLandTemperaturesByCity.csv")
df.head()
```

```python
filtered_data=df[df['City']=='New York']
filtered_data =filtered_data[['dt','AverageTemperature']]

print(filtered_data.head())
```

```python
import matplotlib.pyplot as plt

df['City'] = 'New York'
filtered_data = df[df['City'] == 'New York']
filtered_data = filtered_data.dropna(subset=['AverageTemperature'])

filtered_data['dt'] = pd.to_datetime(filtered_data['dt'])
filtered_data['Year'] = filtered_data['dt'].dt.year
annual_data = filtered_data.groupby('Year')['AverageTemperature'].agg(['min', 'max', 'mean']).reset_index()
plt.figure(figsize=(8, 5))
plt.scatter(annual_data['Year'], annual_data['mean'], color='hotpink', label='Avg Temp', alpha=0.7)
plt.scatter(annual_data['Year'], annual_data['min'], color='skyblue', label='Min Temp', alpha=0.7)
plt.scatter(annual_data['Year'], annual_data['max'], color='indigo', label='Max Temp', alpha=0.7)

plt.title('Annual Temperature Trend for New York', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
